```python
import pandas as pd
import numpy as np
```

```python
data1 = np.random.randn(10,3)
data2 = np.random.randn(7,3)
```

```python
df1
```

```python
df2
```

```python
df3 = pd.concat([df1,df2], ignore_index = True)
df3
```

```python
for i in df3.Col3:
    print(i)
```

```python
# iteritems : to print in the form of keys and values
# iteritems is depricated(removed) in latest versions

# use items instead of iteritems
for key, value in df3.items():
    print(key, value)
```

```python
# iterrows for index and its values
for index, value in df3.iterrows():
    print(index,value)
```

```python
# Clean missing values or NaN values
df
```

```python
# fillna : filling NaN values
df.fillna(0)
```

```python
# Methods to fill NaN values
# Forward method => pad/ffill : fills same as the next row
# Backward method => bfill/backfill : fills same as the previous row
```

```python
df.fillna(method = 'pad')

# FutureWarning: DataFrame.fillna with 'method' is deprecated and will raise in a future version. Use obj.ffill() or obj.bfill() instead.
#   df.fillna(method = 'pad')
```

```python
df.ffill()
```

```python
df.backfill()

# FutureWarning: DataFrame.backfill/Series.backfill is deprecated. Use DataFrame.bfill/Series.bfill instead
#   df.backfill()
```

```python
df.bfill()
```

```python
df.fillna(method = 'bfill')
```

```python
df.ffill(axis = 1)
```

```python
df
```

```python
df = df.bfill()
df
```

```python
# dropna() : used only when we have a very very vvvvvv huge data
df.dropna()
```

```python
df.dropna(axis = 1)
```

```python
# Replacing missing values
df.replace({-0.587215:100, 2.355557:200})
# The df is showing a limited value of the number and we are giving a 
# simple number i.e., -0.587215 is not in the df but -0.587215... is 
# continued with some other values

# To replace such values, its better to round off the values 
# and then replace them
```

```python
df5 = pd.DataFrame({'Apples' : [50,95,105,75], 'Grapes' : [90,23,10,56]})
```

```python
df5
```

```python
df5.replace({50:90, 99:245})
```

```python
df5.replace({95:60.7, 99:245.9})
# The datatype of whole column changes as the inserted values is float
```

```python
df5
```

```python
# df5.Apples.loc[2] = 500  --=> isn't supporting this
# df5['Apples'][2] = 500  =>error

# `df.loc[row_indexer, "col"] = values`
df5.loc[2,'Apples'] = 500
df5.loc[2,'Grapes'] = 200
df5.loc[3,'Apples'] = 50
df5
```

```python
# Handling the duplicate values
df5.duplicated()
```

```python
df5.duplicated(subset = ['Apples'])
```

```python
# Removal of duplicates => drop_duplicates()
df5.drop_duplicates(subset = ['Apples'])
```
