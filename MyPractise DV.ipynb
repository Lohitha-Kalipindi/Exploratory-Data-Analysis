```python
# To get only points
plt.plot(x,y,'o')
# The third parameter is the symbot of the plot points (MATKERS)

# MARKERS
# o : disc
# s : square
# * : star
# p : pentagon
# d : diamond etc
```

```python
plt.plot(x,y,'*')
```

```python
# Types of lines (Reference of lines)
# solid line : -
# dotted line : :
# dashed line : --
# dashed/dotted line : -.
```

```python
# Format string (fmt)
# syntax : marker | line | color
plt.plot(x,y,'d:g')
```

```python
# Reference of colors for basic colors (refer fro others on google)
# r : red
# g : green
# b : blue
# y : yellow
# w : white
# k : black
```

```python
plt.plot(y,'*--k')
# Here, we didn't give the x-axis 
# so it automatically takes the values from 0 (see ticks)
```

```python
# Changing the size of the marker (ms)
plt.plot(y, marker = '*', ms = 25)
```

```python
# Changing the color of the edge of marker (mec)
plt.plot(y, marker = '*', ms = 20, mec = 'r' )
```

```python
# Changing the color of the marker face (mfc)
plt.plot(y, marker = '*', ms = 20, mec = 'r', mfc = 'y' )
```

```python
# We can also give the codes to colors (hex codes)
```

```python
# Linestyle(ls)
plt.plot(y, ls = 'dotted')
# or
plt.plot(y, linestyle = 'dashed')
# or
plt.plot(y, ls = ':')
# or
plt.plot(y, linestyle = '--')
```

```python
# Line width (lw)
plt.plot(y, ls = ':', color='#333', linewidth = '10')
# or
plt.plot(y, ls = ':', color='#333', lw = '10')
```

```python
# Plotting multiple lines in a single graph
plt.plot(y)
```
